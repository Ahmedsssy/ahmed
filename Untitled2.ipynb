```python
import cv2, os, random
import matplotlib.pyplot as plt

def show_samples(split, n=4):
    img_dir = f"FYP-Shoplift-1/{split}/images"
    imgs = random.sample(os.listdir(img_dir), n)

    plt.figure(figsize=(8,4))
    for i, img_name in enumerate(imgs):
        img = cv2.imread(os.path.join(img_dir, img_name))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.subplot(1, n, i+1)
        plt.imshow(img)
        plt.title(f"{split}")
        plt.axis("off")
    plt.show()

show_samples("train")
show_samples("valid")
show_samples("test")
```

```python
import cv2, os, random
import matplotlib.pyplot as plt
import numpy as np

img_dir = "FYP-Shoplift-1/train/images"
img_name = random.choice(os.listdir(img_dir))
img = cv2.imread(os.path.join(img_dir, img_name))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Resize
img_resized = cv2.resize(img, (640,640))

# Grayscale
gray = cv2.cvtColor(img_resized, cv2.COLOR_RGB2GRAY)

# Equalization
equalized = cv2.equalizeHist(gray)

# Blur
blur = cv2.GaussianBlur(equalized, (5,5), 0)

# Canny Edge
edges = cv2.Canny(blur, 50, 150)

# Brightness & Contrast adjustment
def adjust_brightness_contrast(image, brightness=30, contrast=30):
    # Convert to float
    img_float = image.astype(np.float32)
    # Apply contrast
    img_contrast = img_float * (contrast/127 + 1) - contrast
    # Apply brightness
    img_bright = img_contrast + brightness
    # Clip to valid range
    img_bright = np.clip(img_bright, 0, 255).astype(np.uint8)
    return img_bright

bright_contrast = adjust_brightness_contrast(img_resized, brightness=40, contrast=40)

# عرض الصور
plt.figure(figsize=(15,8))

plt.subplot(2,3,1)
plt.title("Original")
plt.imshow(img); plt.axis("off")

plt.subplot(2,3,2)
plt.title("Gray")
plt.imshow(gray, cmap="gray"); plt.axis("off")

plt.subplot(2,3,3)
plt.title("Equalized")
plt.imshow(equalized, cmap="gray"); plt.axis("off")

plt.subplot(2,3,4)
plt.title("Blur")
plt.imshow(blur, cmap="gray"); plt.axis("off")

plt.subplot(2,3,5)
plt.title("Edges")
plt.imshow(edges, cmap="gray"); plt.axis("off")

plt.subplot(2,3,6)
plt.title("Brightness & Contrast")
plt.imshow(bright_contrast); plt.axis("off")

plt.show()
```
